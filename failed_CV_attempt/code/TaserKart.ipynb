Download model with date and time in name

```python
import datetime
from shutil import copy
from google.colab import files

# Get date and time
now = datetime.datetime.now()
date = str(now).split()[0] + '_' + str(now.strftime("%X"))

#copy('TaserKart-main/outputs/model.pth', 
#     'TaserKart-main/outputs/model_'+ date +'.pth')

# Compress outputs folder
from zipfile import ZipFile
import os
  
def get_all_file_paths(directory):
  
    # initializing empty file paths list
    file_paths = []
  
    # crawling through directory and subdirectories
    for root, directories, files in os.walk(directory):
        for filename in files:
            # join the two strings in order to form the full filepath.
            filepath = os.path.join(root, filename)
            file_paths.append(filepath)
  
    # returning all file paths
    return file_paths   
directory = 'TaserKart-main/outputs'
file_paths = get_all_file_paths(directory)
print('Following files will be zipped:')
for file_name in file_paths:
  print(file_name)

# writing files to a zipfile
with ZipFile('outputs_zipped_'+ date +'.zip','w') as zip:
  # writing each file one by one
  for file in file_paths:
    zip.write(file)

#files.download('TaserKart-main/outputs/model_'+ date +'.pth') 
files.download('outputs_zipped_'+ date +'.zip') 
```
